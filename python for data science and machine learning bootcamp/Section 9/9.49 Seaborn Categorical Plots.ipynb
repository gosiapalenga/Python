```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std ,errwidth=0)
```

```python
ax = sns.countplot(x='sex', data=tips)  # counting occurences
ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(10,7))

sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')

# Interpretation: clients on Fridays who are non smokers tend to have a larger bill
```

```python
# Violinplot shows distribution

plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
plt.figure(figsize=(10,6))

sns.stripplot(x='day', y='total_bill',data=tips, jitter=True, hue='sex', dodge=True)
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(x='day', y='total_bill', data=tips, hue='time')
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```
