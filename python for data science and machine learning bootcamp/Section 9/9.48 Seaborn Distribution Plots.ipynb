```python
import numpy as np
import pandas as od
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Univariate Distribution - histogram
```

```python
sns.displot(tips['total_bill'], kde=False, bins=40)
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', multiple='stack', kind='kde')
```

```python
sns.displot(data=tips,x='total_bill', hue='day', multiple='stack', kind='kde')
```

```python
# jointplot - two distribution plots plus a scatter plot
```

```python
sns.jointplot(x='total_bill' ,y='tip' ,data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.rugplot(tips['total_bill'])  # similar to histogram
```

```python
# KDE line: Kernel Density Estimation
```

```python
ax = sns.barplot(x=tips['sex'], y=tips['tip'], errwidth=0)

ax.bar_label(ax.containers[0])
```

```python
ax = sns.barplot(x=tips['day'], y=tips['tip'], errwidth=0)

ax.bar_label(ax.containers[0])
```
