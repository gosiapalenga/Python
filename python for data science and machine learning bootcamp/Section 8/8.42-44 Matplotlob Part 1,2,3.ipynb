```python
# Single plot

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='green', alpha=0.8, lw=3, ls='-', 
        marker='o', markersize=12, markerfacecolor='brown', markeredgewidth=1, markeredgecolor='black')  label='x label')
ax.set_title('x')

ax.plot(y,x, label='y label')
ax.set_title('y')

ax.legend(loc=0)
```

```python
# Multiple plots

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# np.linspace(start,stop,N)
# np.linspace(10,25,10)  10 numbers between 10 and 25

x = np.linspace(0,5,11)
x
```

```python
y = x **2
y
```

```python
# Functional method
```

```python
plt.plot(x,y,'r')

plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Plt title')
```

```python
# Multiplots  --> plt.subplot
```

```python
plt.subplot(1,2,1)  # number of rows, columns and the plot number we are reffering too
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method
```

```python
# Create figure object and call method of this object. 

fig = plt.figure()  # blank canvas

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])   # left, right, width and height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
# Subplots
```

```python
fig, axes= plt.subplots(nrows=3,ncols=3)  # fig.add_axes are included

plt.tight_layout() # fixes plots overlaps
```

```python
fig, axes= plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes are arrays

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
# Figure size

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```
