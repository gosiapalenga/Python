```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Discrete Random Variables following Bernoulli distribution

```python
from scipy.stats import bernoulli
```

```python
data = bernoulli.rvs(size=10000, p=0.4)
sns.distplot(data)
plt.xlabel("Beurnoulli")
plt.ylabel("Freuency")
plt.show()
```

# Discrete Random Variables following Binominal distribution

```python
from scipy.stats import binom
```

```python
data = binom.rvs(n=10, p=0.5, size=10000)
sns.distplot(data)
plt.xlabel("Binominal")
plt.ylabel("Freuency")
plt.show()
```

# Discrete Random Variables following Geometric distribution

```python
from scipy.stats import geom
```

```python
data=geom.rvs(p=0.5, size=10000)
sns.distplot(data)
plt.xlabel("Geometric")
plt.ylabel("Freuency")
plt.show()
```

# Continuous Random Variable following Uniform Distribution

```python
from scipy.stats import uniform

n = 1000
a = 0
b = 10
data = uniform.rvs(size=n, loc=a, scale=b)
sns.distplot(data, bins=100)
plt.xlabel("Uniform")
plt.ylabel("Frequency")
plt.show()
```

# Continuous Random Variable following Exponential Distribution

```python
from scipy.stats import expon

n = 1000
a = 0
b = 1
data = expon.rvs(size=n, loc=a, scale=b)
sns.distplot(data, bins=100)
plt.xlabel("Exponential")
plt.ylabel("Frequency")
plt.show()
```
