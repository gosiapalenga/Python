```python
Calculate Measures of Spread: Range, Variance, Std
```

```python
def get_range(x):
    x_range = np.max(x) - np.min(x)
    return x_range


# def get_variance(x):
#     x_mean = np.sum(x) / len(str(x))
#     x_deviation = [(x - x_mean)**2 for i in x]
#     x_variance = np.sum(x_deviation) / len(str(x))
#     return x_variance


# def get_std(x):
#     x_var = get_variance(x)
#     x_std = np.sqrt(x_var)
#     return x_std


def find_outliers_iqr(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    IQR = Q3 - Q1
    outliers = x[((x < (Q1 - 1.5) * IQR) | (x > (Q3 + 1.5) * IQR))]
    return outliers


def find_outliers_zscore(x):
    x_mean = np.mean(x)
    x_std = get_std(x)
    x_zscore = (x - x_mean) / x_std
    
    return x_zscore
```
