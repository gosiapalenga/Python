```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
language = ['Python', 'R', 'Java', 'HTML', 'C++']
popularity = [22.8, 16, 10, 8, 9]
# colors = ['#FA8F0A', ...]
```

```python
# Pie Chart
```

```python
explode = (0.2, 0,0,0,0) # explode Python slice
plt.pie(popularity, labels=language, autopct= "%1.1f%%", explode = explode)
plt.title("Popularity of Programming Languages")
```

```python
# Histogram
```

```python
values = [82,76,24,67,40,62,75,78,50,71,32,98,89,78,67,72,82,66,87,56,65,85]
```

```python
plt.hist(values, bins = 10, edgecolor = 'black', label = 'Test score data', orientation='horizontal')
plt.title("Histogram for scores")
plt.xlabel("scores")
plt.ylabel("frequency")
plt.legend()
plt.show()
```

```python
# Line plot will give us the overall trend(increasing, decreasing, etc)
```

```python
year = [1960,1970,1980,1990,2000,2010]
pop_china = [667.07, 818.32, 981.24, 1135.20, 1262.60, 1337.70]
pop_india = [449.48, 553.58, 696.78, 870.13, 1053.10, 1231.00]
pop_usa = [180.67, 205.05, 227.23, 249.62, 282.162, 309.34]
```

```python
plt.plot(year, pop_china, label = 'China', color = 'y')
plt.plot(year, pop_india, label = 'India', color = 'b')
plt.plot(year, pop_usa, label = 'USA', color = 'r')
plt.xlabel("Country Names")
plt.ylabel("Population")
plt.legend()
plt.show()
```
