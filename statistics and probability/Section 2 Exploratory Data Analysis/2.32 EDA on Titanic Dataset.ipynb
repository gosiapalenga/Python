```python
data['Initial'] = 0

for i in data:
    data['Initial'] = data['Name'].str.extract('([A-Za-z]+)\.')
```

```python
data.head()
```

```python
pd.crosstab(data['Initial'], data['Sex'])
```

```python
data['Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don']
                        ,['Miss','Miss','Miss','Mr','Mr','Mrs','Mrs','Other','Other','Other','Mr','Mr','Mr'],inplace=True)
```

```python
pd.crosstab(data['Initial'], data['Sex'])
```

```python
data.groupby('Initial')['Age'].mean()
```

```python
# fill na values
data.loc[(data['Age'].isnull()) & (data['Initial'] == 'Mr'), 'Age'] = 33
data.loc[(data['Age'].isnull()) & (data['Initial'] == 'Master'), 'Age'] = 5
data.loc[(data['Age'].isnull()) & (data['Initial'] == 'Miss'), 'Age'] = 22
data.loc[(data['Age'].isnull()) & (data['Initial'] == 'Mrs'), 'Age'] = 36
data.loc[(data['Age'].isnull()) & (data['Initial'] == 'Other'), 'Age'] = 45
```

```python
# convert continuous value(Age) into categorical format
```

```python
data['Age_Cat'] = 0
data.loc[data['Age'] <=16, 'Age_Cat'] = 0
data.loc[(data['Age'] >16) & (data['Age'] <=32), 'Age_Cat'] = 1
data.loc[(data['Age'] >32) & (data['Age'] <=50), 'Age_Cat'] = 2
data.loc[(data['Age'] >50) & (data['Age'] <=70), 'Age_Cat'] = 3
data.loc[data['Age'] >70, 'Age_Cat'] = 4
data.head(10)
```

```python
data['Age_Cat'].value_counts()
```

```python
sns.factorplot("Age_Cat", "Survived", data=data, col="Pclass")

# as the age increases the chance of survival decreses 
```
