```python
import sklearn
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np.random.seed(50)
```

```python
X = 2 * np.random.rand(100,1)    # one column (feature) with 100 rows
y = 4 + 3*X + np.random.randn(100,1)    # noise in the data
```

```python
X[:5]
```

```python
y[:5]
```

```python
plt.plot(X, y, 'bo')
plt.xlabel("X")
plt.ylabel("y")

plt.axis([0,2,0,15])
```

```python
X.shape
```

```python
X_b = np.c_[np.ones((100,1)),X]   # add x0 = 1 for each instance
```

```python
X_b[:10]
```

```python
plt.plot(X_b)
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)   # Normal equation
```

```python
theta_best   # value of co-efficient and intercept
```

```python
X_new = np.array([[0], [3]])
X_new_B = np.c_[np.ones((2,1)), X_new]   # x0=1
y_predict = X_new_B.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-", label='Regression line')
plt.plot(X,y, 'bo')
plt.xlabel("X")
plt.ylabel("y")
plt.axis([0,2,0,15])
plt.legend()
```
