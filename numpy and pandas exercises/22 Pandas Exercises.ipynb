```python
url = (
    'https://storage.googleapis.com/esmartdata-courses-files/'
    'ds-bootcamp/london_bike.csv'
)
```

```python
df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
print(df.head(5))
print()
print(df.tail(5))
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.info()
```

```python
df['hour'] = df['timestamp'].dt.hour
```

```python
df
```

```python
df['month'] = df['timestamp'].dt.month
```

```python
humidity_by_month = df.groupby('month')['hum'].mean().reset_index()
```

```python
humidity_by_month
```

```python
cnt_by_hour = df.groupby('hour')['cnt'].mean().reset_index()
```

```python
cnt_by_hour
```

```python
cnt_by_weekend_hour = df.groupby(['is_weekend', 'hour'])['cnt'].mean().reset_index()
```
