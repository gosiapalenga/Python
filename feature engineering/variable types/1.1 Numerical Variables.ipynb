```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("loan.csv")

data.head()
```

```python
data.info()
```

```python
data.describe()
```

# Continuous Variables

```python
data.plot(kind="scatter", x='customer_id', y='disbursed_amount', figsize=(20,10))
```

```python
fig = data['disbursed_amount'].hist(bins=60, figsize=(20,10))

fig.set_xlabel('Loan amount')
fig.set_ylabel('Number of loans')
```

```python
fig = data['interest'].hist(bins=60, figsize=(20,10))

fig.set_xlabel('Interest rate')
fig.set_ylabel('Number of loans')
```

```python
data.plot(kind='scatter', x='customer_id', y='interest', figsize=(20,10))
```

```python
data['income'].hist(bins=60, figsize=(20,10))
```

```python
fig = data.plot(kind='scatter', x='income', y='customer_id', figsize=(20,10))

fig.set_xlim(0, 500000)
```

# Discrete Variables

```python
data['number_open_accounts'].sort_values().unique()
```

```python
fig = data['number_open_accounts'].hist(bins=50, figsize=(20,10))

fig.set_xlabel('Number of Open Accounts')
fig.set_ylabel('Number of Customers')
```

```python
fig = data.plot(kind='scatter', x='customer_id', y='number_open_accounts', figsize=(20,10))

fig.set_xlim(0, 10000)
```
